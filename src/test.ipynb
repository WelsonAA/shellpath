```python
def traslate(x, y,a=1.45,b=399,c =-1.44 ,d =308):
    return int (a*x+b) ,  int (c*y+d)
```

```python
font                   = cv2.FONT_HERSHEY_SIMPLEX
bottomLeftCornerOfText = (10,500)
fontScale              = 0.5
fontColor              = (255,0,0)
thickness              = 2
lineType               = 1


img = cv2.imread("comp.png")
c = 1

for x,y,_ in givenPoints:
    x,y = traslate(x=x, y=y)

    bottomLeftCornerOfText = (x+12,y+12)
    cv2.putText(img, str(c), bottomLeftCornerOfText, font,fontScale,fontColor,thickness, lineType)
    img = cv2.circle(img, (x, y), 12, (255, 0, 0), 2)
    c+=1
plt.imshow(img)
```
